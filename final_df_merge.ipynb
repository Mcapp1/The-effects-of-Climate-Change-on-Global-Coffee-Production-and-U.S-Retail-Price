## Read in individual DF's

```python
import pandas as pd
import numpy as np
```

### Brazil

```python
brazil_df = pd.read_csv('./data/brazil_df.csv')
brazil_df
```

```python
# change 'Year' column to match 'year'
brazil_df = brazil_df.rename(columns={'Year':'year'})
```

### Colombia

```python
colombia_df = pd.read_csv('./data/colombia_capp.csv')
colombia_df.head()
```

### Ethiopia

```python
ethiopia_df = pd.read_csv('./data/ethiopia/ethiopia_df.csv')
ethiopia_df.head()
```

```python
ethiopia_df.drop(columns='Unnamed: 0', inplace=True)
```

## Merging

```python
braz_colu = pd.merge(brazil_df, colombia_df, how='outer', on='year')
braz_colu.head()
```

```python
braz_colu.info()
```

```python
full_df = pd.merge(braz_colu, ethiopia_df, how='outer', on='year')
full_df
```

```python
full_df.isnull().sum()
```
