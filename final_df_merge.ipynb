## Read in individual DF's

```python
import pandas as pd
import numpy as np
```

### Brazil

```python
brazil_df = pd.read_csv('./data/brazil_df.csv')
brazil_df
```

```python
# change 'Year' column to match 'year'
brazil_df = brazil_df.rename(columns={'Year':'year'})
```

### Colombia

```python
colombia_df = pd.read_csv('./data/colombia_capp.csv')
colombia_df.head()
```

### Ethiopia

```python
ethiopia_df = pd.read_csv('./data/ethiopia/ethiopia_df.csv')
ethiopia_df.head()
```

```python
ethiopia_df.drop(columns='Unnamed: 0', inplace=True)
```

```python
ethiopia_df.year.unique()
```

```python
e_prod = ethiopia_df.iloc[0:30][['year', 'e_prod']]
e_prod.head(3)
```

```python
e_prod['year'] = e_prod['year'].apply(
    lambda x:(x.split("/")[0]))
e_prod.head(3)
```

```python
e_temp_change = ethiopia_df.iloc[30:58][['year', 'e_temp_change']]
e_temp_change.head(3)
```

```python
e_export = ethiopia_df.iloc[58:88][['year', 'e_export']]
e_export.head(3)
```

```python
e_export['year'] = e_export['year'].apply(
    lambda x:(x.split('.')[0]))
```

```python
e_consumption = ethiopia_df.iloc[88:][['year', 'e_consumption']]
e_consumption['year'] = e_consumption['year'].apply(
    lambda x:(x.split("/")[0]))
e_consumption.head(3)
```

```python
prod_temp = pd.merge(e_prod, e_temp_change, how='outer', on='year')
prod_temp.head(3)
```

```python
prod_temp_export = pd.merge(prod_temp, e_export, how='outer', on='year')
prod_temp_export.head(3)
```

```python
ethiopia_df = pd.merge(prod_temp_export, e_consumption, how='outer', on='year')
```

```python
ethiopia_df['year'] = ethiopia_df['year'].astype(int)
```

```python
ethiopia_df.info()
```

## Merging
